```python
import requests
import json

def portuguese_to_english(text: str, api_key: str) -> str:
    """
    Translates Portuguese text to English using Google Cloud Translation API and API key.
    
    Args:
        text: Portuguese text to translate (e.g. "Coloque a figura do curativo")
        api_key: Your Google Cloud API key
    
    Returns:
        Translated English text (e.g. "Place the bandage figure")
    """
    url = f"https://translation.googleapis.com/language/translate/v2"
    
    params = {
        "q": text,
        "source": "pt",
        "target": "en",
        "key": api_key
    }
    
    try:
        # Sending POST request to Google Cloud Translation API
        response = requests.post(url, data=params)
        response_data = response.json()
        
        # Check if the response is valid
        if 'data' in response_data and 'translations' in response_data['data']:
            return response_data['data']['translations'][0]['translatedText']
        else:
            print(f"Error: {response_data}")
            return text  # Return original text in case of error
    except Exception as e:
        print(f"Translation error: {str(e)}")
        return text

# Example usage with your API key
api_key = "AIzaSyAVETX43obXBc0zTfDh1spYfxzSA5Jiu8g"  # Replace with your actual API key
sample_text = "Coloque a figura do estetoscópio na área destacada."
translation = portuguese_to_english(sample_text, api_key)
print(f"Translated: {translation}")
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import Levenshtein

# Load pre-trained SBERT model
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')

def calculate_similarity(text1: str, text2: str):
    """
    Calculates semantic and grammatical similarity between two texts.
    
    Args:
        text1: First text (e.g. Portuguese text)
        text2: Second text (e.g. Translated English text)
    
    Returns:
        semantic_similarity: Cosine similarity of semantic meaning
        grammatical_similarity: Levenshtein distance-based similarity
    """
    # Calculate Semantic Similarity using SBERT
    embeddings1 = model.encode([text1])
    embeddings2 = model.encode([text2])
    
    semantic_similarity = cosine_similarity(embeddings1, embeddings2)[0][0]
    
    # Calculate Grammatical Similarity using Levenshtein distance
    # We normalize it by dividing by the max possible distance (length of the longer string)
    grammatical_similarity = 1 - (Levenshtein.distance(text1, text2) / max(len(text1), len(text2)))
    
    return semantic_similarity, grammatical_similarity



# Apply the function to calculate similarity for each pair of texts
df[['semantic_similarity', 'grammatical_similarity']] = df.apply(
    lambda row: pd.Series(calculate_similarity(row['English'], row['English_GT'])), axis=1)

# Display the results
print(df)
```
